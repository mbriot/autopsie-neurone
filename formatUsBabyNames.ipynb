```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./NationalNames.csv')
```

```python
df.drop_duplicates(subset='Name', keep='first', inplace=True)
```

```python
df.head()
```

```python
df = df.iloc[1:10000]
df = df.drop(['Id','Year','Gender','Count'], 1)
```

```python
def createTarget(name) :
    if len(name) < 3 : return "badBoy"
    target = "badBoy"
    if name[0].lower() == 'a' : target = "goodBoy"
    if name[1].lower() == 'u' : target = "goodBoy"
    if name[2].lower() == 'h' : target = "goodBoy"
    return target
```

```python
df['target'] = df['Name'].apply(lambda name: createTarget(name))
```

```python
df['formatedNames'] = df['Name'].apply(lambda x: x[0:5] if len(x) > 5 else x + 'x' * (5 - len(x)))
```

```python
df = df.drop(['Name'], 1)
```

```python
df = df[['formatedNames', 'target']]
```

```python
df.target.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['formatedNames'], df['target'], test_size=0.25, random_state=5)
```

```python
trainingData = pd.DataFrame(dict(formatedNames = X_train, target = y_train)).reset_index().drop(['index'],1)
validationData = pd.DataFrame(dict(formatedNames = X_test, target = y_test)).reset_index().drop(['index'],1)
```

```python
validationData[validationData['target'] == "badBoy"]
```

```python
trainingData.to_csv('training_data.csv',index=False,header=False)
validationData.to_csv('validation_data.csv',index=False,header=False)
```
